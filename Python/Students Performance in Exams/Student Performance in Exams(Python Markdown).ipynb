First of all, importing the libraries that are required for analysis. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
```

Now we import the CSV file using the below code. 

```python
data = pd.read_csv ('StudentsPerformance.csv')
data.head()
```

Let's move straight to the analysis. We start with describing the data. 

```python
data.describe()
```

We see that there are 1000 observations here. Mean and standard deviations of the variables can be seen in the table above. 

```python
data.groupby('gender').mean()
```

In the above table we see the average math, reading and writing score of male and female. On average females have better scores in reading and writing than males. Let's visualize it.  

```python
sns.set(style='darkgrid')
data.groupby('gender').mean().plot.bar(figsize = (9,6))
plt.show()
```

Using Pairplot in Seaborn Library. We are going to do comparative visualization between male and female scores. 

```python
sns.pairplot(data = data, hue = 'gender')
plt.show()
```

Below we observe Math score distribution on the basis of races (A,B,C & D)

```python
plt.figure(figsize = (10,7))
sns.swarmplot(data = data, x = 'race/ethnicity', y = 'math score', palette = 'magma', hue = 'race/ethnicity')
sns.violinplot(data = data, x = 'race/ethnicity', y = 'math score', palette = 'Pastel2')
plt.show()
```

Now we are going to observe how the variables are correlated with each other using the heatmap plot. We begin by finding the correlation coefficient of variables with respect to each other. And with the help of the correlation table, we make it's heatmap. 

```python
y = data.corr()
y
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(y, annot = True, cmap ='YlOrRd')
plt.show()
```

What is the effect of 'test preparation course' on math, reading and writing scores? We are going to find it now. 

```python
test = data.groupby('test preparation course').mean()
test
```

Students who have completed test preparation course have obtained higher score in all subjects. Now, we will see how test preparation course and math score appear in boxplot.

```python
plt.figure(figsize = (7,5))
sns.boxplot(data = data, x = 'test preparation course', y = 'math score', hue = 'gender')
plt.show()
```
